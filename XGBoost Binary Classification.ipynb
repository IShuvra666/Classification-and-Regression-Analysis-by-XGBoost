```python
# Importing the Libraries

import pandas as pd
import numpy as np
import xgboost as xgb
```

```python
# Importing the Dataset

data = pd.read_csv(r"E:\Python for Data Science and Machine Learning\banking.csv", sep = ";")
pd.set_option('display.max_columns', None)
data.head()
```

```python
data.dtypes
```

```python
# Transforming dep var into 1 and 0

data["y"] = np.where(data["y"] == "yes", 1, 0)
```

```python
dataset_categorical = data.select_dtypes(exclude = "number")
dataset_categorical.head()
```

```python
dataset_numeric = data.select_dtypes(include = "number")
dataset_numeric.head()
```

```python
dataset_categorical = pd.get_dummies(data = dataset_categorical, drop_first = True)
dataset = pd.concat([dataset_numeric, dataset_categorical], axis = 1)
dataset.head()
```

```python
y = dataset.loc[:, "y"]
X = dataset.drop("y",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2,random_state=1502)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
xgbc = XGBClassifier()
print(xgbc)
```

```python
classifier =XGBClassifier()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier,
X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
